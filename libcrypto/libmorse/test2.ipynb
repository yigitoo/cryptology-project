```python
from funcs import *
import sounddevice as sd

def image2wav(image='output.png', out='output.wav'):
    img = Image.open(image)
    ar = np.asarray(img)
    ar = ar.flatten()
    ar2 = []
    ar3 = []
    for i in range(int(len(ar)/3)):
        nt = []
        nt.append(ar[i])
        nt.append(ar[i+1])
        nt.append(ar[i+2])
        ar2.append(nt)
    ar2 = np.array(ar2)
    for i in range(len(ar2)):
        #print(rgb2wve(ar2[i]))
        ar3.append(rgb2wve(ar2[i]))
    return ar3


arwv = image2wav('output.png')
arwv = np.array(arwv, dtype=np.uint8)
sd.play(arwv)
```
